### METODI MONTE CARLO 


Monte carlo è una classe di metodi computazionali basati sul campionamento casuale per ottenere risultati numerici.
La differenza tra simulazione e metodi MC è che la prima è la trascrizione in codici di processi stocastici, mentre  MC è la soluzione tramite metodi probabilistici di problemi non probabilistici. 
Come si può osservare dai grafici la deviazione standard diminuisce al variare dei blocchi.

<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
<span style="color:green">MEDIA A BLOCCHI </span>:
Calcolare l'incertezza statistica, cioè la deviazione standard della media, richiede una stima della varianza $\sigma^2$ : 
        $$\sigma^2_A := \langle (A-\langle A\rangle)^2\rangle = \langle A^2 \rangle -\langle A\rangle^2 $$
La stima del risultato ($A$ generico) nei metodi Monte Carlo è ottenuta tramite $M$ "steps" MC. Dividendo questi "steps" in $N$ blocchi ($M/N$ "steps" per ogni blocco) si può ottenere una stima di $A$ per ogni di questi. 
Chiamiamo $A_i$ with $i=1,N$ e di conseguenza  $A^2_i$ with $i=1,N$.
Si ottiene: 
$$ \langle A^2 \rangle \simeq \frac{1}{N} \sum_{i=1}^N A^2_i \quad \quad \langle A\rangle^2 \simeq \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 $$
    e l'incertezza statistica di $N$ :
    $$\frac{\sigma}{\sqrt{N-1}} \simeq \sqrt{\frac{1}{N-1} \left[ \frac{1}{N} \sum_{i=1}^N A^2_i - \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 \right]} $$
