### Metodo dell'inversione della distribuzione 

*Definizione :* Data un distribuzione di probabilit√† 1D $p(x)$ definita su $-\infty<x<+\infty$ la *funzione  di distribuzione cumulativa* $F(x)$ √® definita come :
$$ F(x) = \int_{-\infty}^{x} dz \, p(x')dx'. $$

Trovata la distribuzione analitca e assegnando $y=F(x) \in [0,1]$ si pu√≤ trovare i valori di x estratti secondo la distribuzione richiesta.
$F(x)$ deve essere invertibile analiticamente per utilizzare questo metodo.



*Lorentziana :*
    $$ x = \Gamma tan\big[\pi \big(y-\frac{1}{2}\big ) \big]$$
    
*Esponenziale :*
    $$ x = -\frac{1}{\lambda} \ln(1-y)$$
