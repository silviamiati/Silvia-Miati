Simulare l'esperimento di Buffon. La probabilità che l'ago intersecherà una delle linee orizzontali (la cui distanza è $d$) è:
$$ P = \frac{2L}{\pi d}. $$ 
Questo può essere utilizzato per ricavarsi il valore di $\pi$ gettando l'ago: 
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
