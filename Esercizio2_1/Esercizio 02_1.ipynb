### Distribuzione Uniforme

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


y,erry,x = np.loadtxt("dis_uni.dat",unpack=True)


plt.errorbar(x,y,yerr=erry)
plt.xlabel('$block$')
plt.ylabel('$<I>$')
plt.axhline(y=1, color='g')
plt.title('Distribuzione uniforme')
plt.grid(True)



plt.show()
```

### Distribuzione non uniforme

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y,erry,x = np.loadtxt("dis_non.dat",unpack=True)


plt.errorbar(x,y,erry)
plt.xlabel('$block$')
plt.ylabel('$<I>$')
plt.axhline(y=1, color='g')
plt.title('Distribuzione non uniforme')
plt.grid(True)


plt.show()
```

### Importance Sampling


L'utilizzo dell'important sampling comporta, come si vede dal grafico, una notevole diminuzione sulla varianza.
La funzione scelta che approssima meglio il coseno nel range tra $[0,1]$ è $1-x$ (correttamente normalizzata).
