<span style="color:red">3D Random Walks (RW) su un reticolo cubico e nel continuo:

1. In un reticolo cristallino con $a=1$. Si è utilizzato un passo ugale a $a$ (lungo una delle tre dimensioni); 
2. Nel continuo. Si è utilizzato un passo uguale ad $a$ (lungo una direzione random).

<span style="color:red"> Disegnare $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ per entrambi i RWs</span>,con la loro relativa incertezza statistica, come funzione di $i\in [0,10^2]$.


### Random Walk Reticolo di passo $a=1$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y,erry,x = np.loadtxt("rw.dat",unpack=True)

plt.errorbar(x,y,yerr=erry)
plt.xlabel('$passi$')
plt.ylabel('$\sqrt{<|{r}_{N}^{2}|>_{RW}}$')
plt.title('Random Walk reticolo')



plt.show()
```
