### Esercizio 03


Calcolare a  $t=0$ **via Monte Carlo** il "European call-option price", $C[S(0),0]$, e "put-option price", $P[S(0),0]$:

1. <span style="color:red"> Campionando  **direttamente** asset price finale $S(T)$</span> per $GBM(r,\sigma^2)$
2. <span style="color:red"> Campionando in maniera discreta il cammino $GBM(r,\sigma^2)$ dell' asset price</span> 
dividendo $[0,T]$ in $10^2$ intervalli di tempo: $S(0)$, $S(t_1)$, $S(t_2)$, ... fino a $S(t_{100}=T=1)$
    

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y,erry,x = np.loadtxt("S_directcall.dat",unpack=True)


plt.errorbar(x,y,yerr=erry)
plt.xlabel('$block$')
plt.ylabel('$<C>$')
plt.title('European call-option prices, $C[S(0),0]$ (direct) ')
plt.grid(True)

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

y,erry,x = np.loadtxt("S_directput.dat",unpack=True)

plt.errorbar(x,y,yerr=erry)
plt.xlabel('$block$')
plt.ylabel('$<P>$')
plt.title('European call-option prices, $P[S(0),0]$ (direct) ')
plt.grid(True)

plt.show()
```
