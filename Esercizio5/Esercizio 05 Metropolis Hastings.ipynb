### Catene di Markov

La catena di Markov è un insieme di variabili casuali $X_{1},X_{2},...$ tale che $X_{i+1}$ dipenda solo dalla variabile precedente, e non dalla storia della catena. Si è interessati in particolare a queste catene in quanto convergono sempre a una distribuzione di equilibrio che sia unica. La condizione di unicità è soddisfatta se si ha ergodicità, cioè se le probabilità $p_{i}(x)$ convergono a $\pi(x)$, che è la distribuzione di equilibrio, quando $\lim_{i \to \inf} p_{i}(x) = \pi(x)$. Le variabili saranno le probabilità a posteriori. Le tecniche MCMC permettono di mappare numericmente la probabilità a posteriori in casi anche molto difficili. L'idea è che tramite un processo iterativo si crei una collezione di stati nello spazio dei parametri in cui i singoli punti vengono aggiornati tramite l'algoritmo di M-H. Questo algoritmo crea una catena di punti tale che seguano la distrivbuzione a posteriori.

In un insieme canonico l'aspettazione rispetto alla posterior della statistica bayesiana equivale all'aspettazione rispetto alla distribuzione canonica nella meccanica statistia.
